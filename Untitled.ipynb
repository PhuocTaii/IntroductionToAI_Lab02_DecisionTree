```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, train_test_split
from sklearn import tree
from sklearn.utils import Bunch
import seaborn as sns
import graphviz 

# get data from file .csv
attributes = ['parents', 'has_nurs', 'form', 'children', 'housing', 'finance', 'social', 'health', 'class']
data = pd.read_csv('nursery.data.csv', names = attributes)

unique_values = data[attributes[-1]].unique().astype(str)
# map data
for col in data.columns:
    dict_map = {}
    unique = data[col].unique()
    for i in range(len(unique)):
        dict_map[unique[i]] = i + 1
    data[col] = data[col].map(dict_map)

# split data
X_train = data.iloc[:, :-1]
y_train = data.iloc[:, -1]


# 1.2.1
# 40/60
feature_train, feature_test, label_train, label_test = train_test_split(X_train, y_train, test_size=0.6, random_state=50, shuffle = True)
for i in attributes[:-1]:
    
    count_train = feature_train[i].value_counts()
    count_test = feature_test[i].value_counts()
    count_original = data[i].value_counts()
    
    count_df = pd.DataFrame({
        'Original': count_original,
        'Train': count_train,
        'Test': count_test
    }).reset_index()
    
    melted_count_df = pd.melt(count_df, id_vars='index', var_name='Dataset', value_name='Count')

    
    
    plt.figure(figsize=(6, 4))
    sns.barplot(x='index', y='Count', hue='Dataset', data=melted_count_df)
    plt.title(f'Distributions of attribute {i} in all the data sets')
    plt.xlabel(i)
    plt.ylabel('Total count')
    plt.xticks(rotation=0)
    plt.legend()
    plt.show()
60/40
feature_train, feature_test, label_train, label_test = train_test_split(X_train, y_train, test_size=0.6, random_state=50, shuffle = True)
80/20
feature_train, feature_test, label_train, label_test = train_test_split(X_train, y_train, test_size=0.8, random_state=50, shuffle = True)
20/80
feature_train, feature_test, label_train, label_test = train_test_split(X_train, y_train, test_size=0.2, random_state=50, shuffle = True)

```

```python
# 1.2.2
custom_dataset = Bunch(data=feature_train, target=label_train, feature_names=attributes[:-1], target_names=['class'])
X, y = custom_dataset.data, custom_dataset.target
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth = 3)
clf = clf.fit(X, y)



dot_data = tree.export_graphviz(clf, out_file = None,
                        feature_names=attributes[:-1],
                        class_names=unique_values,
                        filled=True, rounded=True,
                        special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
